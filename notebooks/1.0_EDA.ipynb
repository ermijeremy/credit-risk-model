#### Correcting data types & Sort

```python
df['TransactionStartTime'] = pd.to_datetime(df['TransactionStartTime'])
df = df.sort_values(by='TransactionStartTime')
```

### Outlier detection

```python
isolation_forest = IsolationForest(n_estimators=100, contamination='auto', random_state=42)

value = df['Value'].values.reshape(-1, 1)
isolation_forest.fit(value)

prediction = isolation_forest.predict(value)
```

```python
plt.scatter(range(len(value)), value, c=prediction, cmap='bwr')
plt.title('Outlier Detection for Value')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```

```python
amount = df['Amount'].values.reshape(-1, 1)
isolation_forest.fit(amount)

prediction = isolation_forest.predict(amount)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(range(len(amount)), amount, c=prediction, cmap='bwr')
plt.title("Outlier Detection for Amount")
plt.xlabel("index")
plt.ylabel("value")
plt.show()
```
